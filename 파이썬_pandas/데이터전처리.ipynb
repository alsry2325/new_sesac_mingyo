```python
df2 = pd.read_csv("Data02.csv")
df2["배송시작일"]   

pd.to_datetime(df2["배송시작일"]) #한글이 섞여있을때는 바로 처리할수가 없다


```

```python
def date_function(x):
    return x[:10]    #배송시작일이라는 데이터가 한줄씩 들어가면서  0번째 부터 9글짜까지 있는 글자 하나만 인덱싱해서 가져온다


#apply() 함수를 이용하자 시리즈데이터에서
df2["배송시작일(dateime)"]   = pd.to_datetime(df2["배송시작일"].apply(date_function))
df2["배송완료일(dateime)"]   = pd.to_datetime(df2["배송완료일"].apply(date_function))
df2.head()

#배송이 얼마나 걸렸는지 완료일에서 시작일을 빼면 걸린시간이 나옴
df2["배송시간"] = df2["배송완료일(dateime)"] - df2["배송시작일(dateime)"]
df2.head()  # 결과 배송시간 컬럼이 새로 추가

df2["배송시간"].describe()   #평균적으로 4일정도 걸린다  47일 걸린것도 았다 이런식으로 분석가능
```

```python
df2["주문일"]   #정수형 

pd.to_datetime(df2["주문일"])  # 이상하게 출력하게됨 그래서

def date_function2(x):
    result = str(x)   #정수형으로는 인덱스를 구할수없기때문에 문자형으로 바꾼다음 인덱스를 구한다
    return result[:4] + "-" + result[4:6] + "-" + result[6:8]

df2["주문일"].apply(date_function2)
```
